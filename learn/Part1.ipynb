```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from matplotlib import font_manager

# 第三方字体，将宋体中的数字字母替换为 Times New Roman 字体
font_path = 'D:/__File/FontLoc/tnw+simsun.ttf'
font_manager.fontManager.addfont(font_path)
property = font_manager.FontProperties(fname=font_path)

config = {
    "mathtext.fontset": 'stix',
    "font.family": 'sans-serif',  # sans-serif/serif/cursive/fantasy/monospace
    "font.sans-serif": property.get_name(),  # 'Simsun'宋体
    "font.size": 13,
    "axes.unicode_minus": False,  # 用来正常显示负号
}
plt.rcParams.update(config)
```

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4], [1,2,3,4]);
```

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots()
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple plot')
ax.legend()
plt.show()
```

### 通用绘图模板


```python
# step1 准备数据
x = np.linspace(0, 2, 100)
y = x**x

# step2 设置绘图样式，可具体绘图使具体调整，可省略
mpl.rc('lines', linewidth=4, linestyle='-.')

# step3 定义布局
fig, ax = plt.subplots()

# step4 绘制图像
ax.plot(x, y, label='special')

# step5 添加标签、文字和图例
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple plot')
ax.legend();
```
