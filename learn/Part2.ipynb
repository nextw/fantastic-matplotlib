<h5>绘制lines</h5>
<ol>
<li>plot方法绘制</li>
<li>Line2D对象绘制</li>
</ol>

```python
x = range(0, 5)
y1 = [2, 5, 7, 8, 10]
y2 = [3, 6, 8, 9, 11]

fig, ax = plt.subplots()
ax.plot(x, y1)
ax.plot(x, y2)
print(ax.lines);
```

```python
x = range(0, 5)
y1 = [2, 5, 7, 8, 10]
y2 = [3, 6, 8, 9, 11]

fig, ax = plt.subplots()
lines = [Line2D(x, y1), Line2D(x, y2, color='orange')]
for line in lines:
    ax.add_line((line))
ax.set_xlim(0, 4)
ax.set_ylim(2, 11);
```

<h4>errorbar绘制误差折线图<h4>
绘制errorbar


```python
fig = plt.figure()
x = np.arange(10)
y = 2.5 * np.sin(x / 20 * np.pi)
yerr = np.linspace(0.05, 0.2, 10)
plt.errorbar(x, y+3, yerr=yerr, fmt='o-', ecolor='r', elinewidth=2);
```
